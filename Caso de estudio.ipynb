<p style="text-align: justify;">Consider the simple RC circuit shown in Fig(a). The capacitor $C$ is assumed to be initially uncharged. At $t = 0,$ the switch connecting the dc voltage source $V_0$ to the RC circuit is closed. Find the voltage $v(t)$ across the $R$ for $t\ge 0.$
</p>
![example1-19](./Figure1-40Haykin.png)

<b>Solution:</b>

   Por ley de Kirchhoff: $$x(t) = V_c(t) + V_R(t)$$
Ley de Ohm
$$i=\frac{dq}{dt}$$
Donde:
$$$$$$ V_c(t)=\frac{q(t)}{C};$$$$$$
$$V_R = Ri = R\frac{dq(t)}{dt}$$
$$$$$$ \frac{1}{R}V_R(t) = \frac{dq(t)}{dt}$$$$$$
$$ \frac{1}{R}V_R(t) = \frac{1}{C}\frac{dV_c(t)}{dt}$$
$$$$$$\frac{1}{RC}\frac{dV_R(t)}{dt} = V_c(t)$$$$$$
Ecuacion diferencial

$$$$$$V_0U(t)=\frac{1}{RC}\frac{dV_R(t)}{dt}+V_R(t)$$$$$$


### Solución mediante factor integrante
$$$$$$V_0U(t) = \frac{1}{RC}\frac{dV_R(t)}{dt} + V_R(T)$$$$$$

$$RC*V_OU(t) = \frac{dV_R(t)}{dt} + RC*V_R(t)$$

$$$$$$μ = e^{\int_{0}^{t}RCdt}dt=e^{tRC}$$$$$$

$$e^{tRC}RC*V_0U(t) = e^{tRC}\frac{dV_R(t)}{dt}+e^{tRC}RC*V_R(t)$$

$$$$$$RC\int_{}^{}V_0U(t)e^{tRC}dt = \int_{0}^{t}\frac{d(e^{tRC}*V_R(t))}{dt}$$$$$$

$$RC*V_0U(t)\frac{e^{tRC}}{RC}+K = e^{tRC}V_R(t)$$

$$$$$$V_0U(t)*e^{tRC}+K=e^{tRC}V_R(t)$$$$$$

$$V_0U(t)\frac{e^{tRC}}{e^{tRC}}+ K\frac{1}{e^{tRC}} = V_R(t)$$

$$$$$$V_R(t) = V_0U(t) + K(e^{-tRC})$$$$$$

Para t=0

$$0 = V_0U(t) + K(e^{0RC})$$

$$$$$$0 = V_0U(t) + K(1)$$$$$$

$$K = -V_0U(t)$$

$$$$$$V_R(t) = V_0U(t) - V_0U(t)e^{-tRC}$$$$$$

Solución
$$V_R(t) = V_0U(t)(1-e^{-tRC})$$

## Respuesta al impulso

$$$$$$x_1(t)=\frac{1}{\Delta}u(t+\Delta/2)$$$$$$
$$x_2(t)=\frac{1}{\Delta}u(t-\Delta/2)$$
$$$$$$y_1(t)=\frac{1}{\Delta}\left(1-e^{-RC(t+\Delta/2)}\right)u(t+\Delta/2),\quad x(t)=x_1(t)$$$$$$
$$y_2(t)=\frac{1}{\Delta}\left(1-e^{-RC(t-\Delta/2)}\right)u(t-\Delta/2), \quad x(t)=x_2(t)$$
$$$$$$x_{\Delta}(t)=x_1(t)-x_2(t)$$$$$$
$$y_{\Delta}= \frac{1}{\Delta}\left(1-e^{-RC(t+\Delta/2)}\right)u(t+\Delta/2) - \frac{1}{\Delta}\left(1-e^{-RC(t-\Delta/2)}\right)u(t-\Delta/2)$$
$$$$$$=  \frac{1}{\Delta}\left(u(t+\Delta/2)- u(t-\Delta/2)\right) - \frac{1}{\Delta}\left( e^{-RC(t+\Delta/2)}u(t+\Delta/2) - e^{-RC(t-\Delta/2)}u(t-\Delta/2)\right)$$$$$$
$$\delta{t}=\lim_{\Delta\to 0}x_{\Delta}(t),$$
$$$$$$\frac{d}{dt}z(t)=\lim_{\Delta\to 0} \left\{\frac{1}{\Delta}
\left( z(t+\Delta/2)-z(t-\Delta/2) \right)\right\}.$$$$$$$$
\begin{eqnarray}
h(t)&=&\lim_{\Delta\to 0} x_{\Delta}(t)\\
    &=&\delta(t)-\frac{d}{dt}\left(e^{-tRC}u(t)\right)\\
    &=&\delta(t)-e^{-tRC}\frac{d}{dt}u(t)-u(t)\frac{d}{dt}\left(e^{-tRC}\right)\\
    &=&\delta(t)-e^{-tRC}\delta(t)+RC*e^{-tRC}u(t), \quad x(t)=\delta(t)
\end{eqnarray}$$
$$$$$$h(t)=RC*e^{-tRC}u(t), \quad\quad x(t)=\delta(t).$$$$$$

## Encontrando la respuesta al impulso unitario mediante Transformada de Fourier
$$ V_0U(t) = \frac{1}{RC}\frac{dV_R(t)}{dt}+V_R(t) $$
$$$$$$ x(t) = \frac{1}{RC}\frac{dy(t)}{dt}+y(t)$$$$$$
$$X(jw) = \frac{1}{RC}Y(jw)(jw) + jw $$
$$$$$$ X(jw) = (\frac{1}{RC}jw + 1)(\frac{1}{jw}) $$$$$$
$$H(jw) = \frac{Y(jw)}{X(jw)}$$
$$$$$$ H(jw) = \frac{1}{\frac{1}{RC}jw+1} $$$$$$
$$ H(jw) = \frac{1}{\frac{jw+RC}{RC}}$$
$$$$$$ H(jw) = RC\frac{1}{jw+RC}$$$$$$
$$ H(jw) = \frac{1}{jw + α} \quad\quad h(t) = e^{-αt} U(t)$$
$$$$$$ h(t) = RC (e^{-tRC})U(t) $$$$$$

```python
%pylab inline

from scipy import signal
font={'family':'serif','serif':['Palatino'], 'size' : 14}
matplotlib.rc('font',**font)
matplotlib.rc('text', usetex=True)

def constantes():
    return 1., 2.

RC , T = constantes()

def delta(t):
    return np.piecewise(t, [t==0, t!=0],[1, 0])

def u(t):
    return np.piecewise(t, [t<0, t>=0], [0, 1])

def impulse(t, RC):
    return np.piecewise(t,[t>=0,t<0],[lambda t: RC*np.exp(-t*RC), 0 ])

systemRC = signal.lti([1.,],[RC, 1])
```

```python
systemRC = signal.lti([1.,],[RC, 1])
t,hlti=systemRC.impulse()
figure(figsize=(12,7))
tt=np.linspace(t.min(),t.max(),20)
ht = impulse(tt,RC)
plt.plot(t,hlti,label='sol scipy')
plt.plot(tt,ht,'o',label='sol ED')
plt.xlabel('Tiempo(s)')
plt.ylabel('$h(t)$',fontsize=18)
plt.title('Respuesta al impulso')
plt.legend()
plt.grid()
plt.show()
```

## La respuesta al impulso unitario en el espacio de frecuencias

<b>sympy</b>

```python
import sympy
from sympy.abc import t, tau
R,C, w, w_0 = sympy.symbols('R C w w_0 ')

from sympy.plotting import plot
```

```python
s=sympy.exp(-t*R*C)*R*C
s
```

```python
d=sympy.Integral(s*sympy.exp(-sympy.I*w*t),(t,0, sympy.oo))
d
```

```python
d=d.doit(conds='none')
d
```

```python
d=d.subs(C,1).subs(R,1).doit()
d
```

```python
modd=sympy.functions.Abs(d) 
modd.doit()
```

```python
ws,ds=systemRC.freqresp()
plot(modd,(w,0,10))
plt.plot(ws,abs(ds), label='lti')
plt.xlabel('$w(Hz)$',fontsize=18)
plt.ylabel('$|H(w)|$',fontsize=18)
plt.legend()
plt.grid()
plt.show()
```

```python
argd = sympy.functions.arg(d)
plot(argd, (w,0,10))
plt.plot(ws,np.angle(ds), label='lti')
plt.xlabel('$w(Hz)$',fontsize=18)
plt.ylabel('$arg\{H(w)\}$',fontsize=18)
plt.legend()
plt.grid()
plt.show()
```
